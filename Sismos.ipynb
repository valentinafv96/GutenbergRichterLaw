```python
#Importando librerías

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definiendo listas

file=np.loadtxt('Tabla')   #Estrayendo valores de la tabla

M = file[:,0]              #Magnitudes promedio de los sismos

n = np.log10(file[:,1])    #Número de sismos

N = []                     #Número acumulado de sismos
 
#Ajuste lineal     

b,a = np.polyfit(M,n,1)

print(b,a)
```

```python
x = np.linspace(0,7,30)
y = b*x+a

plt.figure(figsize=(10,6))

plt.plot(M,n,'ko', label='Datos')
plt.plot(x,y,'c',label='$Log_{10}(n) = %.2f x %.2f$' %(a, b))

plt.legend(fontsize=14)

plt.title('\n Ley de Gutemberg-Richter''\n', fontsize=16, fontweight='bold')

plt.xlabel('\n $Magnitud$', fontsize=14)
plt.ylabel('$Log(n)$ \n', fontsize=14)

plt.xlim(0,7)
plt.ylim(0,6)

plt.savefig('GRLaw.png')

plt.show()
```
